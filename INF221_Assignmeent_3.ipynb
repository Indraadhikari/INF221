# Exercise 1 (4 points)
One metric for the productivity of scientists is the h-index. It is defined as follows: For each scientific publication by a scientist, count the number of times it has been cited by other scientists (there are several organisations collecting such statistics). A scientist has an h-index of n if her n-th least cited paper has been cited at least n times. For example, Google Scholar currently reports for me an h-index of 25 at present, i.e., my 25th-least cited publication has been cited (at least) 25 times.

Give an algorithm that determines the h-index of the scientist based on a list of publications and number of citations.
Discuss the worst-case and average-case run-time of your algorithm.

1. Give an algorithm that determines the h-index of the scientist based on a list of publications and number of citations.

## Solution
**Algorithm Pseudo Code:**

```
H-Index(A)
1:  l = A.length
2:  if l == 0
3:      return 0
4:  Sort A in descending order
5:  min = A[l]
6:  if min ≥ l
7:      return l
8:  h = min
9:  while h < A[h + 1]
10:      h = h + 1
11:  return h
```

**Implementation in Python**

````python
def h_index(A):
    if not A:
        return 0
    
    A.sort(reverse=True)
    length = len(A)
    min = A[length-1]

    if min >= length:
        return length
   
    h = min
    while (h < A[h]):
        h = h + 1
    return h

A =[10, 7, 6, 5, 3, 2, 1]
h_index(A)
````

2. Discuss the worst-case and average-case run-time of your algorithm.

## Solution

**Best Case** (non-empty) 
- $min≥l$ : all papers have $≥l$ citations. and $min<l$ but $h=min$
- Returns at Line 7 right after sorting.
- Running time is dominated by sorting (comparison-based sort, $T(n)=Θ(nlog⁡n)$)
- Total $T(n)=Θ(1) + Θ(nlog⁡n) = Θ(nlog⁡n)$\
- *Note: for an empty list, the total run time will be Θ(1).*

**Worst Case**
- $min=0$ and $h=n−1$ (one paper has $0$ citations, all others have $≥n−1$ citations).
- Example: $A=[10,10,10,...,10,0]$
- The while loop runs $n−1$ times.
- The while loop contributes $Θ(n)$, but sorting contributes $Θ(nlog⁡n)$ (comparison-based sort).
- Total $T(n)=Θ(nlog⁡n) + Θ(n) =Θ(nlogn)$

## Exercise 2 (3 points)
Modify the `List-Insert` algorithm for doubly linked lists so that it supports insertion not only at the head of the list, but after an arbitrary element. If the modified algorithm is called as `List-Insert(x, y)`, then y shall be inserted immediately after x. When x = NIL is passed, y shall be inserted at the head. Give your algorithm in pseudocode. Discuss the runtime of your algorithm, taking into account that x gives you direct access to the list element after which to insert.

## Solution

**Pseudo Code**

```
List-Insert(L, x, y)
1:  if x == NIL
2:      y.prev = NIL
3:      y.next = L.head
4:      if L.head != NIL
5:          L.head.prev = y
6:      L.head = y
7:  else
8:      y.prev = x
9:      y.next = x.next
10:     if x.next != NIL
11:         x.next.prev = y
12:     x.next = y
```

**Runtime Analysis**

Only a fixed number of pointer operations are performed in both cases, so the algorithm runs in a constant time: $O(1)$.

## Solution

**Pseudo Code**

```
Delete-Singly(L, k)
1:    if L.head == NIL
2:        error
3:    if L.head.key == k
4:        L.head = L.head.next
5:        return
6:    prev = L.head
7:    curr = L.head.next
8:    while curr != NIL and curr.key != k
9:        prev = curr
10:       curr = curr.next
11:   if curr == NIL
12:        error
13:   prev.next = curr.next
```

**Runtime Analysis**

In the worst case, the while loop costs $O(n)$; otherwise, it costs $O(1)$.

## Exercise 5 (5 points)

You are given a list of intervals $(a_k, b_k)$. These could, e.g., represent times for which meeting rooms are booked. Create an algorithm that reports overlapping intervals. I.e. $L=\{(2,5), (6, 9), (7, 11), (10, 15), (12, 14), (13,17), (19, 21)\}$, for  the algorithm should report that
- (6, 9) overlaps with (7, 11)
- (7, 11) overlaps with (10, 15)
- (10, 15) overlaps with (12, 14) and (13, 17)
- (12, 14) overlaps with (13, 17) 

Each overlap only needs to reported once, i..e, report only that (6, 9) overlaps with (7, 11), but not that (7, 11) overlaps with (6, 9). Discuss the runtime of your algorithm.

## Solution

**Pseudo Code**
```
Find-Overlaps(L)
1:  n = length(L)
2:    for i = 0 to n - 2
3:        for j = i + 1 to n - 1
4.            if L[j].a < L[i].b
5:                report "L[i] overlaps with L[j]"
6:            else
7:                break
```

**Runtime Analysis**

Best case 
- No overlaps, that means the inner loop will break with one step
- so, $O(n)$ run time

Worst case
- All intervals overlap each other, which means the inner loop will go in full scale. In total, because of both loops, it creates a quadratic running cost
- So , $O(n^2)$ run time
