One metric for the productivity of scientists is the h-index. It is defined as follows: For each scientific publication by a scientist, count the number of times it has been cited by other scientists (there are several organisations collecting such statistics). A scientist has an h-index of n if her n-th least cited paper has been cited at least n times. For example, Google Scholar currently reports for me an h-index of 25 at present, i.e., my 25th-least cited publication has been cited (at least) 25 times.

Give an algorithm that determines the h-index of the scientist based on a list of publications and number of citations.
Discuss the worst-case and average-case run-time of your algorithm.

1. Give an algorithm that determines the h-index of the scientist based on a list of publications and number of citations.

## Solution
**Algorithm Pseudo Code:**

```
H-Index(A)
1:  l = A.length
2:  if l == 0
3:      return 0
4:  Sort A in descending order
5:  min = A[l]
6:  if min ≥ l
7:      return l
8:  h = min
9:  while h < A[h + 1]
10:      h = h + 1
11:  return h
```

**Implementation in Python**

````python
def h_index(A):
    if not A:
        return 0
    
    A.sort(reverse=True)
    length = len(A)
    min = A[length-1]

    if min >= length:
        return length
   
    h = min
    while (h < A[h]):
        h = h + 1
    return h

A =[10, 7, 6, 5, 3, 2, 1]
h_index(A)
````

2. Discuss the worst-case and average-case run-time of your algorithm.

## Solution

**Best Case** (non-empty) 
- $min≥l$ : all papers have $≥l$ citations. and $min<l$ but $h=min$
- Returns at Line 7 right after sorting.
- Running time is dominated by sorting (comparison-based sort, $T(n)=Θ(nlog⁡n)$)
- Total $T(n)=Θ(1) + Θ(nlog⁡n) = Θ(nlog⁡n)$\
- *Note: for an empty list, the total run time will be Θ(1).*

**Worst Case**
- $min=0$ and $h=n−1$ (one paper has $0$ citations, all others have $≥n−1$ citations).
- Example: $A=[100,100,100,...,100,0]$
- The while loop runs $n−1$ times.
- The while loop contributes $Θ(n)$, but sorting contributes $Θ(nlog⁡n)$ (comparison-based sort).
- Total $T(n)=Θ(nlog⁡n) + Θ(n) =Θ(nlogn)$
