### Cost Analysis
- Line 2: No. of execution: $n+1$, cost: $c2$
- Line 3: No. of execution: $n$, cost: $c3$
- Line 4: No. of execution: $1$, cost: $c4$
- Line 5: No. of execution: $1$, cost: $c5$

The total runtime is then: 

\begin{align}
T(n) &= c2(n+1) + c3n + c4 + c5 \\
&= (c2 +c3)n + c2+c4+c5 \\

\end{align}

- $n/2$ comparison.

We often need to sort collections of strings, for examples names in a phonebook. Advanced sorting algorithms will often convert all the strings to numbers first (each string to a different number) and then sort by comparing those numbers (known as sort keys).

Why would it be an advantage to use numbers as sort keys instead of sorting the strings directly?
Consider the method used to convert strings to numbers: which condition must the numbers created fulfill?
Can you suggest an algorithm for converting strings into numbers to be used as sort keys? Assume that each letter in a string is 1 byte long, that numbers are 8 bytes long and that all strings are at most 8 letters long.

## Solution

### Why would it be an advantage to use numbers as sort keys instead of sorting the strings directly?

- Total operations without converting to number for Insertion sort
(Where $L$ is the leangth of the strings and $n$ is zise of the array.):
$$O(L.n^2)$$ 
- Total operations with converting to number for Insertion sort:
$$O(L.n) +O(n^2)$$

- $L.n^2$ is greater than $L.n + n^2$ for large $n$, therefore it would be an advantage to use numbers as sort keys instead of sorting the strings directly.

- For other sorting algorithms (like MergeSort), total operations without converting to number
(Where $L$ is the leangth of the strings ):
$$O(L.nlog⁡n)$$
- Total operations with converting to number:
$$O(L.n) + O(nlogn)$$

- $L.nlog⁡n$ is greater than $L.n + nlogn$ for large $n$, therefore it would be an advantage to use numbers as sort keys instead of sorting the strings directly.

- **Example**
    - For L = 5, n = 8
    - Total cost without numeric key = $ 5*8*log⁡(8) = 5*24 = 120$
    - Total cost using numeric key = $ 5*8 + 8*log⁡(8) = 40 + 24 = 64$


### Consider the method used to convert strings to numbers: which condition must the numbers created fulfill?

**It must be fulfill the alphabetical Condition:** \
If string $S_1$ comes before string $S_2$ in alphabetical (lexicographical) order, then the number generated for $S_1$ must be less than the number generated for $S_2$.
$$key(S_1) < key(S_2)$$


Can you suggest an algorithm for converting strings into numbers to be used as sort keys? Assume that each letter in a string is 1 byte long, that numbers are 8 bytes long and that all strings are at most 8 letters long.

### Pseudocode

```
1:  StringToSortKey(s)
2:  key = 0
3:  for i = 1 to length(s)
4:      key = key * 256
5:      key = key + ASCII(s[i])
6:  return key
```

````python
def StringToSortKey(s):
    #print(ord(""))
    key = 0
    for i in range (0,len(s)):
        key = key * 256
        key = key + ord(s[i])
    return key

def insertion_sort(A):
    for j in range(1, len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i] > key:
            A[i + 1] = A[i]
            i = i - 1
        A[i + 1] = key

def KeyToString(n):
    chars = []
    
    while n > 0:
        c = n % 256          
        chars.append(chr(c))  
        n //= 256            
    return ''.join(reversed(chars))

#s = "Markdown"
#print(StrintToSortKey(s))

S = ["ABCDEFGH","ZZZZZZZZ", "ABCDEFGA", "AAAAAABA", "AAAABAAA", "AAAAAAAA", "AAAAAAAB"]
print("Before sorting",S)

A = []
for i in range(0,len(S)):
    key = StringToSortKey(S[i])
    A.append(key)

#print(A)

insertion_sort(A)
#print(A)

S_sorted = []
for i in range (0, len(A)):
    string = KeyToString(A[i])
    S_sorted.append(string)

print("After sorting",S_sorted)
````

## Exercise 4 (2 points)

Assume that for inputs of size $n$, the runtime for insertion sort is  $T_I(n) = 16n²$ and for merge sort $T_M(n) = 512n log n$. 
For which values of $n$ is insertion sort faster than merge sort?


## Solution

````python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 1001)

T_I = 16 * n**2
T_M = 512 * n * np.log2(n)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, T_I, label="Insertion Sort: 16n²")
plt.plot(n, T_M, label="Merge Sort: 512n log₂ n")

# crossover
# approximate crossover
for i in range(2, 1001):
    if 16*i*i >= 512*i*np.log2(i):
        crossover = i
        break

plt.axvline(x=crossover, color='red', linestyle='--', label=f"Crossover = {crossover}")

plt.xlabel("n")
plt.ylabel("Runtime")
plt.title("Insertion Sort vs Merge Sort Runtime")
plt.legend()
plt.grid(True)

plt.show()
````

We have experimented for total runtime on both sorting methods with $n = 1$ to $1000$, and plot on the graph. From the experiment, we can see that the insertion sort is faster than merged sort for $1<n<256$.

## Solution

**1. Develop a recursive algorithm to solve this task.**
### Pseudocode

```
1:  recursive_min(A,n)  
2:  if n == 1
3:      return A[1]
4:  res = recursive_min(A, n-1)
5:  if A[n] < res
6:      return A[n]
7:  else
8:      return res

```

**2. Draw a recursion tree for the algorithm and estimate the runtime $T(n)$.**

Level $0$:        $T(n)$   


Level $1$:       $T(n-1)$          
Level $2$:       $T(n-2)$           

...              ...

Level $n-1$:      $T(1)$  

All level has same work cost lets assume $c$, then the estimate runtime is
\begin{equation}
T(n) = \sum_{k=1}^n c_i \;
\end{equation}
$$T(n) = c + c + c + ....+ c = cn$$
$$T(n) = cn$$

**3. Find a recurrence equation for $T(n)$.**

The total time $T(n)$ is the sum of the recursive call and the constant-time overhead:
$$T(n) = T(n-1) + c$$

**4. Implement your recursive algorithm in Python and compare its runtime to the loop-based algorithm from the lab session in week 6 using `%timeit`.**

````python
def recursive_min(A,n):
    if n == 1:
        return A[1]
    res = recursive_min(A, n-1)
    if A[n-1] < res:
        return A[n-1]
    return res


def minimum(data):
    min_value = data[0]
    for next_value in data[1:]:
        if next_value < min_value: 
            min_value = next_value
    return min_value
````
