## Exercise 1 (4 points)
Use the master method to give tight asymptotic bounds for the following recurrences if the master method can be applied:\
a) $T(n)=2T(n/4)+1$ \
b) $T(n)=2T(n/4)+\sqrt{n}$ \
c) $T(n)=2T(n/4)+n$ \
d) $T(n)=2T(n/4)+n^2$ 

## Solution

**Solving recurrences using the Master method**

General solution to recurrences of the form:
- $T(n) = a T(n/b) + f(n) \qquad\text{where}\; a\geq 1, b>1, f(n)>0.$

**Case 1:**\
Condition:
- $f(n) = O(n^{\log_b a-\epsilon}) \quad\text{for some}\; \epsilon>0$
- $f(n)$ is polynomially smaller than $n^{\log_b a}$

Solution:
$T(n) = \Theta(n^{\log_b a})$

**Case 2:**\
Condition:
- $f(n) = \Theta(n^{\log_b a})$
- $f(n)$ grows as fast as $n^{\log_b a}$ 

Solution:
$T(n) = \Theta(n^{\log_b a}\lg n)$

**Case 3:**\
Condition:
- $f(n) = \Omega(n^{\log_b a+\epsilon}) \quad\text{for some}\; \epsilon>0$ and regularity condition
- $a f(n/b) \leq c f(n) \quad\text{for some}\; c< 1\;\text{and all large $n$}$
- $f(n)$ is polynomially greater than $n^{\log_b a}$
- Regularity condition always fulfilled if $f(n)=n^k$ is polynomial

Solution:
$T(n) = \Theta(f(n))$

**a) $T(n)=2T(n/4)+1$**
- $T(n) = 2T(n/4) + 1 \implies a=2, b=4, f(n) = 1$
- Compare $n^{\log_4 2}$ to $n^0$
- We have $\log_4 2 < 0 \implies 1/2 - \epsilon = 0 \implies$ Case 1
- $T(n) = \Theta(n^{\log_4 2}) \implies \Theta(n^{1/2})$


**b) $T(n)=2T(n/4)+\sqrt{n}$**
- $T(n) = 2T(n/4) + \sqrt{n} \implies a=2, b=4, f(n) = \sqrt{n}$
- Compare $n^{\log_4 2}$ to $n^{1/2}$
- We have $\log_4 2 = 1/2 \implies$ Case 2
- $T(n) = \Theta(n^{1/2} \lg n)$

**c) $T(n)=2T(n/4)+n$**
- $T(n) = 2T(n/4) + n \implies a=2, b=4, f(n) = n$
- Compare $n^{\log_4 2}$ to $n^1$
- We have $\log_4 2 < 1 \implies 1/2 + \epsilon = 1 \implies$ Case 3
- $T(n) = \Theta(n)$

**d) $T(n)=2T(n/4)+n^2$**
- $T(n)=2T(n/4)+n^2 + \Theta(n^2) \implies a=5, b=2, f(n) = n^2$
- Compare $n^{\log_4 2}$ to $n^2$
- We have $\log_4 2 < 2 \implies 1/2 + \epsilon = 2 \implies$ Case 3
- $T(n) = \Theta(n^2)$

## Exercise 2 (4 points)
The following algorithm to create random permutations of array A, but should **never** return the original ordering of A. The algorithm is similar to the one we investigated in class. But it is also correct, i.e., will it generate all possible permutations of A (except the original A itself) with equal probability? Assume that all elements of A are unique.
```
Permute-Without-Identity(A)
  n = A.length
  for i = 1 to n-1
    swap A[i] with A[Random(i+1, n)]
```
Hint: Begin by experimenting with small arrays A. No more than three elements are needed to see how the algorithm behaves. 

## Solution

**Analysis of Permute-Without-Identity(A)** \

First Iteration, $i=1$ \
$A[1]$ is always swap with $A$[``Random`` $(2,n)$]. So $A[1]$ is always change with some other element. That means the original combination will never be generated. 

Probability of possible permutation with small array: \
For example, $A = [x,y,z]$ so, $n=3$
- $i = 1, \implies$ swaping between $x$ and $y$ or $z$. So, it can be either *Case1:* $[y,x,z]$ or *Case2:* $[z,y,x]$ with equal probability $1/2$.
- $i = 2, \implies$ swaping between $x$ and $z$ on *Case1* and swaping between $y$ and $x$ on *Case2* depending on the privious iteration. So, it can be either $[y,z,x]$ or $[z,x,y]$.
- So, finally the result can be $[y,z,x]$ or $[z,x,y]$ with probability $1/2$ each.
- It can only produce two results among five total non-identity permutation for a array with three element array.

````python
import random
from collections import Counter

def Permute_Without_Identity(A):
    n = len(A)
    A = A[:]
    for i in range(n-1):
        r = random.randint(i+1,n-1)
        A[i], A[r] = A[r], A[i]
    return A

A = ["x", "y", "z"]
num_trials = 5000
outcome_counts = Counter()

for _ in range(num_trials):
    result = Permute_Without_Identity(A)
    outcome_counts[tuple(result)] += 1 

print(f"Results over {num_trials} trials:\n")
print(f"{'Outcome':<20} {'Count':<10} {'Probability'}")
for outcome, count in sorted(outcome_counts.items()):
    probability = count / num_trials
    print(f"{str(list(outcome)):<20} {count:<10} {probability:.4f} ({probability*100:.2f}%)")
````
